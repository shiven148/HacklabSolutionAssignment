```python
import matplotlib.pyplot as plt
import cv2
import glob

# Path to test results
test_result_path_highres = '/content/yolov5/runs/detect/exp/labels/*.txt'
test_result_path_augmented = '/content/yolov5/runs/detect/exp2/labels/*.txt'

# List all result files
test_result_files_highres = glob.glob(test_result_path_highres)
test_result_files_augmented = glob.glob(test_result_path_augmented)

# Function to visualize detection results
def visualize_results(image_path, label_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    with open(label_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            class_id, x_center, y_center, width, height, confidence = map(float, line.strip().split())
            h, w, _ = img.shape
            x1 = int((x_center - width / 2) * w)
            y1 = int((y_center - height / 2) * h)
            x2 = int((x_center + width / 2) * w)
            y2 = int((y_center + height / 2) * h)
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img, f'Person: {confidence:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Display a few high resolution test results
for result_file in test_result_files_highres[:5]:
    image_file = result_file.replace('labels', 'images').replace('.txt', '.jpg')
    visualize_results(image_file, result_file)

# Display a few augmented test results
for result_file in test_result_files_augmented[:5]:
    image_file = result_file.replace('labels', 'images').replace('.txt', '.jpg')
    visualize_results(image_file, result_file)
```
